# Работа с Pandas и CSV
Основные операции с данными

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Загрузка данных

```python
df = pd.read_csv('sample_data.csv')
df
```

## 2. Информация о данных

```python
print('Размер таблицы:', df.shape)
print('\nТипы данных:')
print(df.dtypes)
```

```python
df.describe()
```

## 3. Фильтрация данных

```python
# Сотрудники из IT отдела
it_dept = df[df['department'] == 'IT']
it_dept
```

```python
# Зарплата выше 80000
high_salary = df[df['salary'] > 80000]
high_salary
```

## 4. Группировка и агрегация

```python
# Средняя зарплата по отделам
df.groupby('department')['salary'].mean()
```

```python
# Количество сотрудников по городам
df['city'].value_counts()
```

## 5. Визуализация

```python
# Средняя зарплата по отделам (столбчатая диаграмма)
df.groupby('department')['salary'].mean().plot(kind='bar', color=['#3498db', '#e74c3c', '#2ecc71'])
plt.title('Средняя зарплата по отделам')
plt.xlabel('Отдел')
plt.ylabel('Зарплата')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Распределение по городам (круговая диаграмма)
df['city'].value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.title('Распределение сотрудников по городам')
plt.ylabel('')
plt.show()
```

## 6. Сохранение результатов

```python
# Сохранить отфильтрованные данные в новый CSV
it_dept.to_csv('it_employees.csv', index=False)
print('Файл it_employees.csv сохранён!')
```
