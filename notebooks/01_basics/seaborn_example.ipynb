# Красивые графики с Seaborn
Примеры визуализации данных

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Устанавливаем стиль
sns.set_theme(style='whitegrid')
print('Seaborn готов!')
```

## 1. Загрузка встроенного датасета

```python
# Датасет о чаевых в ресторане
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
tips.info()
```

## 2. Распределения

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Гистограмма с KDE
sns.histplot(tips['total_bill'], kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Распределение суммы счёта')

# Гистограмма по категориям
sns.histplot(data=tips, x='total_bill', hue='time', kde=True, ax=axes[1])
axes[1].set_title('Счёт: обед vs ужин')

plt.tight_layout()
plt.show()
```

## 3. Категориальные графики

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Box plot
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', ax=axes[0])
axes[0].set_title('Box Plot: счёт по дням')

# Violin plot
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, ax=axes[1])
axes[1].set_title('Violin Plot')

# Bar plot
sns.barplot(data=tips, x='day', y='tip', hue='sex', ax=axes[2])
axes[2].set_title('Средние чаевые по дням')

plt.tight_layout()
plt.show()
```

## 4. Scatter plots

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', size='size', sizes=(50, 200))
plt.title('Чаевые vs Счёт')
plt.show()
```

```python
# Scatter с регрессией
plt.figure(figsize=(10, 6))
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha': 0.5})
plt.title('Чаевые vs Счёт (с линией регрессии)')
plt.show()
```

## 5. Pair Plot — матрица графиков

```python
sns.pairplot(tips, hue='sex', height=2.5)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

## 6. Тепловая карта корреляций

```python
# Корреляция числовых признаков
numeric_cols = tips.select_dtypes(include=[np.number])
corr = numeric_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Матрица корреляций')
plt.show()
```

## 7. FacetGrid — несколько графиков

```python
g = sns.FacetGrid(tips, col='time', row='smoker', height=4)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='day')
g.add_legend()
g.figure.suptitle('FacetGrid: разбивка по времени и курению', y=1.02)
plt.show()
```

## 8. Разные стили Seaborn

```python
styles = ['whitegrid', 'darkgrid', 'white', 'dark', 'ticks']
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for ax, style in zip(axes, styles):
    with sns.axes_style(style):
        sns.histplot(tips['tip'], ax=ax, color='coral')
        ax.set_title(f'style: {style}')

plt.tight_layout()
plt.show()
```

## 9. Цветовые палитры

```python
palettes = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for ax, palette in zip(axes.flat, palettes):
    sns.barplot(data=tips, x='day', y='total_bill', hue='sex', palette=palette, ax=ax)
    ax.set_title(f'palette: {palette}')
    ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
